The goal of this project is to develop a machine learning algorithm to identify abnormal ECG waveforms. This model will be an anaomaloy detection algorithm using an autoencoder. The autoencoder will seek to reconstruct the ECG waveform based on encoded information from the input waveform. The model will be trained to accuratley reconstruct a normal ECG waveform, but not an abnormal ECG waveform. Because of this, when the autoencoder sees an abnormal ECG waveform its reconstruction error will be significantly higher than a typical normal ECG waveform. This will allow autoencoder reconstruction error, or mean absolute error, to be used to identify normal vs. abnormal waveforms. The error value can be adjusted to adjust the sensitivity of the model. This type of model could be used for real-time detection of abnormal ECG waveforms, which could be useful for monitoring of patient's cardiac health.

The first step is to install/import all necessary python libraries and download the dataset used for this project

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, losses
from tensorflow.keras.models import Model
```

```python
# Download the dataset
dataframe = pd.read_csv('http://storage.googleapis.com/download.tensorflow.org/data/ecg.csv', header=None)
raw_data = dataframe.values
dataframe.head()
```

```python
# The last element contains the labels
labels = raw_data[:, -1]

# The other data points are the electrocadriogram data
data = raw_data[:, 0:-1]

train_data, test_data, train_labels, test_labels = train_test_split(
    data, labels, test_size=0.2, random_state=21
)
```

```python
min_val = tf.reduce_min(train_data)
max_val = tf.reduce_max(train_data)

train_data = (train_data - min_val) / (max_val - min_val)
test_data = (test_data - min_val) / (max_val - min_val)

train_data = tf.cast(train_data, tf.float32)
test_data = tf.cast(test_data, tf.float32)
```

```python
train_labels = train_labels.astype(bool)
test_labels = test_labels.astype(bool)

normal_train_data = train_data[train_labels]
normal_test_data = test_data[test_labels]

anomalous_train_data = train_data[~train_labels]
anomalous_test_data = test_data[~test_labels]
```

```python
plt.grid()
plt.plot(np.arange(140), normal_train_data[0])
plt.title("A Normal ECG")
plt.show()
```

```python
plt.grid()
plt.plot(np.arange(140), anomalous_train_data[0])
plt.title("An Anomalous ECG")
plt.show()
```

```python
class AnomalyDetector(Model):
    def __init__(self):
        super(AnomalyDetector, self).__init__()
        self.encoder = tf.keras.Sequential([
          layers.Dense(32, activation="relu"),
          layers.Dense(16, activation="relu"),
          layers.Dense(8, activation="relu")])

        self.decoder = tf.keras.Sequential([
          layers.Dense(16, activation="relu"),
          layers.Dense(32, activation="relu"),
          layers.Dense(140, activation="sigmoid")])

    def call(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

autoencoder = AnomalyDetector()

autoencoder.compile(optimizer='adam', loss='mae')
```

```python
history = autoencoder.fit(normal_train_data, normal_train_data, 
          epochs=20, 
          batch_size=512,
          validation_data=(test_data, test_data),
          shuffle=True)
```

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
```

```python
encoded_imgs = autoencoder.encoder(normal_test_data).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()

plt.plot(normal_test_data[0], 'b')
plt.plot(decoded_imgs[0], 'r')
plt.fill_between(np.arange(140), decoded_imgs[0], normal_test_data[0], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

```python
encoded_imgs = autoencoder.encoder(anomalous_test_data).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()

plt.plot(anomalous_test_data[0], 'b')
plt.plot(decoded_imgs[0], 'r')
plt.fill_between(np.arange(140), decoded_imgs[0], anomalous_test_data[0], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

```python
reconstructions = autoencoder.predict(normal_train_data)
train_loss = tf.keras.losses.mae(reconstructions, normal_train_data)

plt.hist(train_loss[None,:], bins=50)
plt.xlabel("Train loss")
plt.ylabel("No of examples")
plt.show()
```

```python
#Select original threshold value
threshold = np.mean(train_loss) + np.std(train_loss)
print("Threshold: ", threshold)
```

```python
reconstructions = autoencoder.predict(anomalous_test_data)
test_loss = tf.keras.losses.mae(reconstructions, anomalous_test_data)

plt.hist(test_loss[None, :], bins=50)
plt.xlabel("Test loss")
plt.ylabel("No of examples")
plt.show()
```

```python
def predict(model, data, threshold):
    reconstructions = model(data)
    loss = tf.keras.losses.mae(reconstructions, data)
    return tf.math.less(loss, threshold)

def print_stats(predictions, labels):
    print("Accuracy = {}".format(accuracy_score(labels, preds)))
    print("Precision = {}".format(precision_score(labels, preds)))
    print("Recall = {}".format(recall_score(labels, preds)))
```

```python
preds = predict(autoencoder, test_data, threshold)
print_stats(preds, test_labels)
```

Depending on the use case of this model, it may be necessary to adjust these precision/recall values. Because abnormal ECG waves can be a sign of a serious cardiac event, the cost associated with misclassifying an abnormal wave may be significantly higher than misclassifying a normal wave (or in other words, that false negatives are significantly more serious than false positives because false negatives could result in patients not recieving medical assistance that they need while false positives will only result in a patient recieving unnecessary medical attention). 

While the overall accuracy of this original model is around 94%, its recall is only around 90%. This means that only about 90% of abnormal ECG waves were correctly identified by the model, with the remaining 10% misclassified as normal ECG waves. This misclassification rate may be much too high. In order to address this, the threshold error value will be increased. This is done below, and the accuracy metrics are recalculated. 

```python
#Adjust threshold to improve recall

threshold = np.mean(train_loss) + 2*np.std(train_loss)
print("Threshold: ", threshold)
```

```python
preds = predict(autoencoder, test_data, threshold)
print_stats(preds, test_labels)
```

After increasing the threshold error value the overall model accuracy decreased, but the model recall increased. This means that more abnormal waveforms were correctly classified, but more normal waveforms were also misclassified as abnormal. As previously stated, this may be desirable for an abnormal ECG detection algorithm because of the potential severity associated with abnormal waveform misclassification compared to normal waveform misclassification. This shows how the threshold value can be used to adjust the classification accuracy of each of the two classes. The final threshold value will depend on the acceptable accuracy, precision and recall values of this type of model.
